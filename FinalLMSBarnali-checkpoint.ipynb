# Mini Project on Python basic Library mangement System #

```python

print("Welcome to Library Management System\n")

# Book data stored in dictionary
library = {
    "Python Basics": {
        "author": "Test",
        "available": True,
        "issued_to": ""
    },
    "Data Science": {
        "author": "Test1",
        "available": False,
        "issued_to": "Radha"
    }
}

# Login credentials
credentials = {
    "admin": "admin123",
    "user1": "user1234"
}

# Login Function

def login():
    attempts = 3
    while attempts > 0:
        user = input("Enter username: ")
        pwd = input("Enter password: ")

        if user in credentials and credentials[user] == pwd:
            print("\nLogin Successful!\n")
            return True
        else:
            attempts -= 1
            print("Wrong username or password")
            print("Attempts left:", attempts, "\n")

    print("Login failed. Program terminated.")
    return False


# View All Books

def view_books():
    print("\nLibrary Books:")
    for book in library:
        if library[book]["available"]:
            status = "Available"
        else:
            status = "Issued"

        print("Book Name :", book)
        print("Author    :", library[book]["author"])
        print("Status    :", status)
        print("----------------------")
    print()


# Add New Book

def add_book():
    name = input("Enter book name: ")
    author = input("Enter author name: ")

    if name in library:
        print("Book already exists!\n")
    else:
        library[name] = {
            "author": author,
            "available": True,
            "issued_to": ""
        }
        print("Book added successfully!\n")


# Issue Book

def issue_book():
    name = input("Enter book name to issue: ")

    if name not in library:
        print("Book not found!\n")
    elif library[name]["available"] == False:
        print("Book already issued!\n")
    else:
        student = input("Enter student name: ")
        library[name]["available"] = False
        library[name]["issued_to"] = student
        print("Book issued successfully!\n")


# Return Book

def return_book():
    name = input("Enter book name to return: ")

    if name not in library:
        print("Book not found!\n")
    elif library[name]["available"] == True:
        print("Book is already available!\n")
    else:
        library[name]["available"] = True
        library[name]["issued_to"] = ""
        print("Book returned successfully!\n")

# View Issued Books

def view_issued_books():
    print("\nIssued Books List:")
    found = False

    for book in library:
        if library[book]["available"] == False:
            found = True
            print("Book Name :", book)
            print("Issued To :", library[book]["issued_to"])
            print("----------------------")

    if found == False:
        print("No books are issued.")

    print()


# Main Menu

def menu():
    while True:
        print("1. View Books")
        print("2. Add Book")
        print("3. Issue Book")
        print("4. Return Book")
        print("5. View Issued Books")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            view_books()
        elif choice == "2":
            add_book()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            view_issued_books()
        elif choice == "6":
            print("Thank you for using LMS")
            break
        else:
            print("Invalid choice! Try again.\n")


# Program Start

if login():
    menu()
```
